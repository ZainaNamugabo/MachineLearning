```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('/content/nike_shoes_sales.csv')
df = df.drop('images', axis=1)
df = df.drop('description', axis=1)
df = df.set_index('product_id')
df.head()
```

```python
plt.figure(figsize=(6,5))
sns.histplot(df["sale_price"], bins=20, kde=True)
plt.title("Distribution of Sale Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.savefig('sale_price_histplot.png')
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.boxplot(df['sale_price'])
plt.title("Boxplot of Sale Prices")
plt.ylabel("Price")
plt.savefig('sale_price_boxplot.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["sale_price"], y=df["rating"])
plt.title("Sale Price vs. Rating")
plt.xlabel("Sale Price")
plt.ylabel("Rating")
plt.savefig('sale_price_rating_scatterplot.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df[["rating", "reviews"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation between Ratings and Reviews")
plt.show()
```
